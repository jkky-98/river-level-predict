```python
def get_data_api(url):
    response = requests.get(url)
    content = response.text
    xml_obj = bs4.BeautifulSoup(content,'lxml-xml')

    obsdh = xml_obj.findAll('obsdh')
    rwl = xml_obj.findAll('rwl')
    iqty = xml_obj.findAll('iqty')
    dambsarf = xml_obj.findAll('dambsarf')
    tdqty = xml_obj.findAll('tdqty')
    edqty = xml_obj.findAll('')

    date = [] # 시간
    wl = [] # 수위
    income = [] # 유입량
    rain = [] # 강우량
    tdqty_l = [] # 총방류량

    for i in obsdh: #시간데이터
        date.append(i.text)
    for i in rwl: #수위데이터
        wl.append(i.text)
    for i in iqty: #유입량데이터
        income.append(i.text)
    for i in dambsarf: #강수량데이터
        rain.append(i.text)
    for i in tdqty: #총방류량
        tdqty_l.append(i.text)


    sub_dict = {'date' : date,
                'wl' : wl,
                'iqty' : income,
                'rain' : rain,
                'tdqty' : tdqty_l}
    df = pd.DataFrame(sub_dict)
    return df
```
